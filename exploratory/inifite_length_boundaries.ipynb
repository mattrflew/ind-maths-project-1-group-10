# Purpose
We have an infinite vein length assumption. For our numerical methods that means we need to assume there is sufficient length behind and past our area of interest such that our solution is not affected by the boundary conditions.

This leads to the main question: **What is a "*sufficient*" length for our purposes?**

To answer this question we will rerun the numerical solution over many distances and observe when our solution begins to get affected.

For computational efficiency, it is in our interest to model the minimum amount of length behind and past our area of interest. 

# Set up

*Note: The results of this are likely also dependent on our choice for $\Delta x$*

We define our *area of interest* as:

$$
0 \leq x \leq 13
$$

Let $L_1$ be the distance behind and $L_2$ be the distance past the area of interest respectively such that:

$$

-\infty < L_1 < 0 \\

13 < L_2  < \infty \\

$$

Use our analytical solution, $C_0 = 8.73 \times 10^{-3}$, such that $C_f = 1.00 \times 10^{-3}$ to see when our final result, $C_f$, starts getting affected.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from functions import *
import time
```

```python
# Model parameters
C0 = 8.73e-3

Nt_gaps = 10000    # number of timesteps
T = 60             # final time 

Nx_spaces = 1000;   #Â number of spaces in x
```

Note: This following cell will take a while to run

```python
x_heart = 13

L = 30

L1 = np.linspace(-L, 0, 10)
L2 = np.linspace(x_heart, x_heart + L, 11)

# Initialize a matrix to store our results
Cf_all = np.zeros((L1.shape[0], L2.shape[0]))

t0 = time.time()
for i,l1 in enumerate(L1):
    print(f'i: {i+1} of {L1.shape[0]}')
    for j,l2 in enumerate(L2):
        print(f'j: {j+1} of {L2.shape[0]}')
        Cf_all[i,j], _, _, _ = forward_euler(Nt_gaps, Nx_spaces, l1, l2, C0=C0, T=T, D=0.1, v=0.2, b0=0, bL=0, x_heart=x_heart)
t1 = time.time()
```

```python
print(f'Time to execute {t1-t0} s')
```
