# Compare the Analytical Solution to the Numerical Methods

### The Analytical Solution

$$
C(\eta, t) = \frac{C_0}{2} [ \mathrm{erf}(\frac{\eta}{2\sqrt{Dt}}) - \mathrm{erf}(\frac{\eta - L}{2\sqrt{Dt}})] \\

\eta = x - vt \\ 

L = 1


$$






```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def eta_func(x, v, t):
    return x - v*t

def erf_func(eta, D, t):
    return math.erf(eta/(2*np.sqrt(D*t)))

def concentration(C0, x, t, v=0.2, D=0.1, L=1):
    x_points = len(x)
    t_points = len(t)
    
    U = np.zeros((x_points, t_points))
    
    for i,xx in enumerate(x):
        for j,tt in enumerate(t):
            eta = eta_func(xx, v, tt)
            U[i, j] = C0/2 * (erf_func(eta, D, tt) - erf_func(eta-L, D, tt))

    return U 
```

```python
x = np.linspace(0,13, 100)
t = np.linspace(0.00000001,60, 100)

C0 = 8.73e-3 # Analytical result

U = concentration(C0, x, t)
```
